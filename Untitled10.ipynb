#Import python library for visualization

```
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```
def bar_chart(feature):
  survived = train[train['Survived']==1][feature].value_counts()    # 생존조건+퓨쳐
  dead = train[train['Survived']==0][feature].value_counts()      # 사망조건+퓨쳐
  df = pd.DataFrame([survived, dead])
  df.index = ['Survived', 'Dead']
  df.plot(kind='bar', stacked=0, figsize=(10, 5))
```

```
bar_chart('Embarked')
```
